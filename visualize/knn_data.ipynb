```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = torch.load('knn_data.pth')
selected_num = data['selected_num']
```

```python
y = np.asarray(selected_num)
x = np.arange(len(y))
```

```python
plt.plot(x, y)  
num = y[-1]/100
indexes = [0,10,20,30,60, 110]
for index in indexes:
    plt.text(x[index]+50, y[index]-300, f'({x[index]}, {y[index]}/ {y[index]/num:.2f}%)', fontsize=10, ha='right')  
    # plt.figure(figsize=(8, 6))  # 设置图表大小为宽度8英寸，高度6英寸 
    s1 = plt.scatter(x[index], y[index], s=25, c='r')  # stroke, colour
plt.title('Batch - Labeled points Line')  # 设置标题  
plt.xlabel('Batch')  # 设置 x 轴标签  
plt.ylabel('Labeled points amount/ ratio')
plt.legend(handles=[s1],labels=['efficiency(batch,amount/ratio)'],loc='best')
plt.plot(x, y) 
```

```python
y_increase = np.copy(y)
num = len(y_increase)
for i in range(num):
    
    y_increase[num-1-i] -= y_increase[num-1-i-1]
        
# print(y_increase[1]-y_increase[0], y_increase[2]-y_increase[1])
index = 40  # 假设要标记第三个点  
indexes = [0,10,20,30,40,50,60, 110]
for index in indexes:
    plt.text(x[index]+40, y_increase[index]+300, f'({x[index]}, {y_increase[index]}/ {y_increase[index]/num:.2f}%)', fontsize=10, ha='right')  
    # plt.figure(figsize=(8, 6))  # 设置图表大小为宽度8英寸，高度6英寸 
    s1 = plt.scatter(x[index], y_increase[index], s=25, c='r')  # stroke, colour
plt.title('Inference Efficiency')  # 设置标题  
plt.xlabel('Batch')  # 设置 x 轴标签  
plt.ylabel('New labeled points amount/ ratio')
plt.legend(handles=[s1],labels=['efficiency(batch,amount/ratio)'],loc='best')
plt.plot(x, y_increase)  
```
